## Step 3: Clean the Data

```python
df=df.drop_duplicates()
```

```python
###Handle missing values
df=df.dropna()
```

```python
df=df.dropna(subset=["model"])
```

```python
## Fill missing values
df.fillna(df["mpg"].mean(),inplace=True)
df.fillna(df["mpg"].median(),inplace=True)
```
